```python
import pandas as pd
import matplotlib.pyplot as plt
url ='1.csv'
dataframe = pd.read_csv(url)
x1=dataframe.SAT
y=dataframe.GPA
plt.scatter(x1,y)
plt.show()
```

```python
import pandas as pd
url ='mtcars.csv'
data=pd.read_csv(url)
data.head(10)
data.shape
```

```python
data.hist()
plt.show()
correlation=data.corr()
print(correlation)
x=data.drop(data.columns[[0,2]],axis=1)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['wt']),y=data['mpg'])
y_prel=model.predict(X=pd.DataFrame(data['wt']))
residual=data['mpg']-y_prel
residual.describe()
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
plt.plot(data['wt'],y_prel,color='blue')
```

```python
model=linear_model.LinearRegression()
predictor=pd.DataFrame([data['wt'],data['wt']**2]).T
model.fit(predictor,y=data['mpg'])
print('Intercept của mô hình bậc 2')
print(model.intercept_)
print('Coeffient của mô hình bậc 2')
print(model.coef_)
print('Độ chính xác của mô hình')
print(model.score(predictor,y=data['mpg']))
```

```python
import numpy as np
range = np.arange(1.5,5.5,0.1)
predictor=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=predictor)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0.7))
plt.plot(range,y_pred,color='blue')
```
